```python
import pandas as pd
import numpy as np
from statsmodels.tsa.arima.model import ARIMA

# Suponha que df_filtrado já está definido

# Inicializar o DataFrame para armazenar todas as previsões simuladas
df_monte_carlo = pd.DataFrame()

# Iterar sobre cada coluna em df_filtrado
for column in df_filtrado.columns:
    # Copiar e preparar os dados
    df = df_filtrado.copy()
    df[column] = pd.to_numeric(df[column], errors='coerce')
    df = df.dropna(subset=[column])

    # Criar o modelo ARIMA
    p, d, q = 1, 2, 3  # Ajuste conforme necessário
    model = ARIMA(df[column], order=(p, d, q))
    model_fit = model.fit()

    # Fazer previsões para os próximos 100 períodos
    forecast_steps = 100
    num_simulations = 10

    # Inicializar o DataFrame para armazenar previsões simuladas da coluna atual
    forecast_data = pd.DataFrame()

    # Gerar simulações de previsões com intervalos de confiança
    np.random.seed(0)  # Para reprodução dos resultados
    for i in range(num_simulations):
        forecast = model_fit.get_forecast(steps=forecast_steps)
        forecast_mean = forecast.predicted_mean
        confidence_intervals = forecast.conf_int(alpha=0.05)  # 95% de confiança

        # Simular previsões dentro do intervalo de confiança
        simulated_forecast = forecast_mean + np.random.uniform(
            confidence_intervals.iloc[:, 1] - forecast_mean,
            confidence_intervals.iloc[:, 0] - forecast_mean,
            size=len(forecast_mean)
        )
        forecast_data[f'{column}_{i+1}'] = simulated_forecast

    # Adicionar os dados simulados ao DataFrame final
    df_monte_carlo = pd.concat([df_monte_carlo, forecast_data], axis=1)

# Exibir o DataFrame com todas as previsões simuladas
print(df_monte_carlo.head())
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.optimizers import Adam

# Carregar os dados
df = df_filtrado.drop(['Date'], axis=1).copy()

# Definir as colunas de entrada e saída
input_columns = df.columns.tolist()
output_columns = ['Receita', 'Lucro Líquido', 'Despesas Operacionais', 'EBITDA', 'Endividamento']

# Preparar os dados de entrada (X) e saída (y)
X = df[input_columns].values[:-1]  # Todos os dados exceto a última linha
y = df[output_columns].values[1:]  # Todos os dados exceto a primeira linha

# Normalizar os dados
scaler_X = MinMaxScaler()
scaler_y = MinMaxScaler()
X_scaled = scaler_X.fit_transform(X)
y_scaled = scaler_y.fit_transform(y)

# Dividir os dados em conjuntos de treino e teste
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y_scaled, test_size=0.2, random_state=42)

# Definir a arquitetura da rede neural
model = Sequential([
    Dense(64, activation='relu', input_shape=(X.shape[1],)),
    Dropout(0.2),
    Dense(32, activation='relu'),
    Dropout(0.2),
    Dense(16, activation='relu'),
    Dense(5)  # 5 neurônios de saída para as 5 variáveis que queremos prever
])

# Compilar o modelo
model.compile(optimizer=Adam(learning_rate=0.001), loss='mean_squared_error')
```

```python
model.summary()
```

```python
# Treinar o modelo
history = model.fit(X_train, y_train, validation_split=0.2, epochs=100, batch_size=32, verbose=1)

# Avaliar o modelo
test_loss = model.evaluate(X_test, y_test)
print(f"Perda no conjunto de teste: {test_loss}")

# Fazer previsões
last_data_point = X_scaled[-1].reshape(1, -1)
prediction_scaled = model.predict(last_data_point)
prediction = scaler_y.inverse_transform(prediction_scaled)

# Imprimir as previsões
for i, col in enumerate(output_columns):
    print(f"Previsão para {col}: {prediction[0][i]:.2f}")

# Plotar o histórico de treinamento
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
plt.plot(history.history['loss'], label='Perda no Treino')
plt.plot(history.history['val_loss'], label='Perda na Validação')
plt.title('Histórico de Treinamento')
plt.xlabel('Época')
plt.ylabel('Perda')
plt.legend()
plt.show()
```

```python
# Função para fazer previsões para múltiplos passos à frente
def predict_multiple_steps(model, initial_data, steps):
    predictions = []
    current_data = initial_data.copy()

    for _ in range(steps):
        prediction = model.predict(current_data.reshape(1, -1))
        predictions.append(prediction[0])

        # Atualizar os dados de entrada com a previsão
        current_data = np.roll(current_data, -5)
        current_data[-5:] = prediction

    return np.array(predictions)

# Fazer previsões para os próximos 5 passos
future_steps = 50
multi_step_predictions_scaled = predict_multiple_steps(model, X_scaled[-1], future_steps)
multi_step_predictions = scaler_y.inverse_transform(multi_step_predictions_scaled)

# Imprimir as previsões de múltiplos passos
for step in range(future_steps):
    print(f"\nPrevisões para o passo {step + 1}:")
    for i, col in enumerate(output_columns):
        print(f"{col}: {multi_step_predictions[step][i]:.2f}")

# Plotar as previsões de múltiplos passos
plt.figure(figsize=(15, 10))
for i, col in enumerate(output_columns):
    plt.subplot(3, 2, i+1)
    plt.plot(range(future_steps), multi_step_predictions[:, i], marker='o')
    plt.title(f'Previsão de {col}')
    plt.xlabel('Passos à frente')
    plt.ylabel('Valor')
plt.tight_layout()
plt.show()
```

```python
multi_step_predictions_df = pd.DataFrame(multi_step_predictions, columns=output_columns)

multi_step_predictions_df['Step'] = np.arange(1, len(multi_step_predictions_df) + 1)

# Reordenar as colunas para que 'Step' seja a primeira
cols = ['Step'] + output_columns
multi_step_predictions_df = multi_step_predictions_df[cols]

# Definir 'Step' como o índice
multi_step_predictions_df.set_index('Step', inplace=True)

# Exibir as primeiras linhas do DataFrame
print(multi_step_predictions_df.head())
```
